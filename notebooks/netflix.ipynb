# Netflix Customer Churn Analysis
### Objective: To identify key drivers of customer attrition and provide data-driven retention strategies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Premium styling
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['font.family'] = 'sans-serif'

# Load dataset
df = pd.read_csv('../datasets/netflix_churn.csv')
print(f"Dataset Loaded: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

## 1. Churn Overview
What is the overall churn rate in our subscriber base?

```python
plt.figure(figsize=(8, 8))
churn_counts = df['churned'].value_counts()
plt.pie(churn_counts, labels=['Retained', 'Churned'], autopct='%1.1f%%', 
        colors=['#2ecc71', '#e74c3c'], explode=(0.05, 0.05), shadow=True, startangle=140)
plt.title('Overall Customer Churn Rate', fontsize=18, fontweight='bold')
plt.show()
```

## 2. Watch Hours and Churn
Does higher engagement (watch hours) correlate with better retention?

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='watch_hours', hue='churned', fill=True, palette='coolwarm', alpha=0.6)
plt.title('Density of Watch Hours: Churned vs Retained', fontsize=16)
plt.xlabel('Monthly Watch Hours')
plt.ylabel('Density')
plt.show()
```

## 3. Subscription Type Impact
Which tier (Basic, Standard, Premium) has the highest loyalty?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='subscription_type', hue='churned', palette='viridis')
plt.title('Churn Count by Subscription Tier', fontsize=16)
plt.show()
```

## 4. Genre Preferences
Are certain content categories more effective at retaining users?

```python
genre_churn = df.groupby('favorite_genre')['churned'].mean().sort_values(ascending=False)
plt.figure(figsize=(14, 6))
sns.barplot(x=genre_churn.index, y=genre_churn.values, palette='magma')
plt.axhline(df['churned'].mean(), color='red', linestyle='--', label='Avg Churn Rate')
plt.title('Churn Rate by Favorite Genre', fontsize=16)
plt.ylabel('Churn Probability')
plt.legend()
plt.show()
```

## 5. Age and Engagement
How does user age relate to watch time and churn?

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df.sample(1000), x='age', y='watch_hours', hue='churned', alpha=0.6)
plt.title('Age vs Watch Hours Correlation', fontsize=16)
plt.show()
```

## 6. Financial Analysis
Correlation between monthly fees and churn behavior.

```python
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='RdBu', fmt='.2f', center=0)
plt.title('Numerical Feature Correlation Matrix', fontsize=16)
plt.show()
```
