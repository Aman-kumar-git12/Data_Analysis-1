# 📺 Netflix Churn: Simple & Powerful Insights
### *Theme: Netflix Red & Deep Black*

This analysis breaks down why customers leave Netflix and identifies the most loyal subscriber segments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Netflix Branding Colors
NETFLIX_RED = '#E50914'
PURE_BLACK = '#000000'
LIGHT_GRAY = '#F5F5F1'

# Setting the 'Black & Red' Premium Theme
plt.style.use('dark_background')
plt.rcParams.update({
    'axes.facecolor': PURE_BLACK,
    'figure.facecolor': PURE_BLACK,
    'grid.color': '#222222',
    'text.color': LIGHT_GRAY,
    'axes.labelcolor': LIGHT_GRAY,
    'xtick.color': LIGHT_GRAY,
    'ytick.color': LIGHT_GRAY,
    'font.size': 12
})

df = pd.read_csv('../datasets/netflix_churn.csv')
print("Analysis ready.")
```

## 1. Retention Pulse
Are we keeping our customers?

```python
plt.figure(figsize=(10, 7))
data = df['churned'].value_counts()
plt.pie(data, labels=['Staying (Retained)', 'Leaving (Churned)'], 
        autopct='%1.1f%%', startangle=140, 
        colors=['#333333', NETFLIX_RED], 
        explode=(0, 0.1), shadow=True)
plt.title('Overall Customer Retention Status', fontsize=20, color=NETFLIX_RED, pad=30, fontweight='bold')
plt.show()
```

## 2. Watch Hours Scatter Analysis
How does engagement relate to age and churn?

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df.sample(800), x='age', y='watch_hours', hue='churned', palette=['#444444', NETFLIX_RED], alpha=0.7, s=100)
plt.title('Scatter Analysis: Age vs Watch Hours', fontsize=18, color=NETFLIX_RED, pad=20)
plt.xlabel('Customer Age', fontsize=14)
plt.ylabel('Monthly Watch Hours', fontsize=14)
plt.legend(['Active', 'Churned'])
plt.show()
```

## 3. Tier Loyalty
Which plan has the most loyal fans?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='subscription_type', hue='churned', palette=['#444444', NETFLIX_RED])
plt.title('Subscriber Count: Active vs Churned by Plan', fontsize=18, color=NETFLIX_RED, pad=20)
plt.show()
```

## 4. Regional Volatility
Where is our churn risk highest?

```python
plt.figure(figsize=(14, 7))
region_churn = df.groupby('region')['churned'].mean().sort_values() * 100
region_churn.plot(kind='barh', color=NETFLIX_RED, edgecolor='white')
plt.title('Churn Rate % by Global Region', fontsize=18, color=NETFLIX_RED, pad=20)
plt.grid(axis='x', alpha=0.2)
plt.show()
```
